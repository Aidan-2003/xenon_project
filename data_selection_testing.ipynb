```python
tmin = 11.5
tail_stop = 200

pS2_times = np.sort(np.asarray(pS2s["time_since_start"], dtype=float))
DE_times = np.asarray(DEs["time_since_start"], dtype=float)

rows = []

for idx, src in enumerate(pS2s):
    t0 = float(src["time_since_start"])

    next_idx = np.searchsorted(pS2_times, t0, side="right")
    next_pS2_dt = np.inf
    if next_idx < len(pS2_times):
        next_pS2_dt = pS2_times[next_idx] - t0

    a = t0 + tmin
    b = t0 + tail_stop

    n_tail = np.sum((DE_times >= a) & (DE_times < b))

    rows.append({
        "time": t0,
        "n_tail_200ms": n_tail,
        "next_pS2_dt_ms": next_pS2_dt,
        "has_next_pS2_within_200ms": next_pS2_dt < 200.0,
        "area": float(src["area"]),
        "n_electron_rec": int(src["n_electron_rec"]),
        "range_50p_area_ms": float(src["range_50p_area"]) / 1e6,
        "subtype": int(src["subtype"]),
        "r": float(src["r"]),
    })

tail_df_2 = pd.DataFrame(rows)

isolated = tail_df_2[tail_df_2["next_pS2_dt_ms"] >= 200.0]
nonisolated = tail_df_2[tail_df_2["next_pS2_dt_ms"] < 200.0]

print("All pS2s:", len(tail_df_2))
print("Fraction <=2, all:", np.mean(tail_df_2["n_tail_200ms"] <= 2))

print("Isolated pS2s:", len(isolated))
print("Fraction <=2, isolated:", np.mean(isolated["n_tail_200ms"] <= 2))

print("Non-isolated pS2s:", len(nonisolated))
print("Fraction <=2, non-isolated:", np.mean(nonisolated["n_tail_200ms"] <= 2))
```

```python
def live_duration_in_region(a, b, pS2s_for_deadtime, s1_times, tmin):
    dead = mod2.build_dead_intervals(
        a, b,
        pS2s_for_deadtime["time_since_start"],
        s1_times,
        tmin
    )
    live = mod2.build_live_intervals(a, b, dead)
    return sum(y - x for x, y in live), live

s1_times = np.asarray(S1s["time_since_start"], dtype=float)

live_ms_list = []

for _, row in tail_df_2.iterrows():
    a = row["time"] + tmin
    b = row["time"] + tail_stop

    live_ms, _ = live_duration_in_region(a, b, pS2s, s1_times, tmin)
    live_ms_list.append(live_ms)

tail_df_2["live_ms_200ms"] = live_ms_list
tail_df_2["live_fraction"] = tail_df_2["live_ms_200ms"] / (tail_stop - tmin)

good_live = tail_df_2[tail_df_2["live_fraction"] > 0.8]

print("Good-live pS2s:", len(good_live))
print("Fraction <=2 among good-live:", np.mean(good_live["n_tail_200ms"] <= 2))
```

```python
yield_summary = tail_df_2.groupby("subtype").agg(
    N=("n_tail_200ms", "size"),
    mean_tail=("n_tail_200ms", "mean"),
    median_tail=("n_tail_200ms", "median"),
    mean_area=("area", "mean"),
    median_area=("area", "median"),
    mean_ne=("n_electron_rec", "mean"),
    median_ne=("n_electron_rec", "median"),
)

yield_summary["tail_per_mean_ne"] = (
    yield_summary["mean_tail"] / yield_summary["mean_ne"]
)

display(yield_summary.sort_values("tail_per_mean_ne"))
```
